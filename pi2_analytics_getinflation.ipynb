##ip2_analytics_getinflation
Recupera el histórico de inflación en los Estados Unidos, para agregar contexto al análisis.

```python
import csv
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#de aquí me traeré los datos de inflación de los states
strUrl = 'https://www.usinflationcalculator.com/inflation/historical-inflation-rates/'
req = requests.get(strUrl)
```

```python
#aquí me traigo la página y guardo la primera tabla en un dataframe. No uso el Id porque la etiqueta <TABLE> de la página no lo tiene 
soup = BeautifulSoup(req.text, 'lxml')
inflationTable = soup.find_all('table')[0]
data_inflation = pd.read_html(str(inflationTable))[0]
```

```python
data_inflation.dtypes
```

```python
data_inflation.shape
```

```python
#Para poderlo comparar con la data de los índices del spx necesito que esté en un formato como:
#fecha | valor, para ello recorreré este primer data frame
lstMeses = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
data_inflation_final=pd.DataFrame([])
for item in range(len(data_inflation)):
    #data_inflation_final['Date']
    for mes in lstMeses:
        fila = {'Date':str(data_inflation.iloc[item]['Year'])+'-'+mes+'-28', 'value': data_inflation.iloc[item][mes]}
        data_inflation_final = data_inflation_final.append(fila,ignore_index=True)
```

```python
data_inflation_final.tail(20)
```

```python
#hay que eliminar los registros vacíos
data_inflation_final.drop(data_inflation_final[data_inflation_final['value'].isnull()].index,axis=0,inplace=True)
```

convertimos la columna 'Date' a formato fecha y lo establecemos como indice

```python
#convierte la fecha de str a datetie
data_inflation_final['Date'] = pd.to_datetime(data_inflation_final['Date'],format='%Y-%b-%d')
#y convierte la columna a indice
data_inflation_final.set_index('Date',inplace=True)
```
